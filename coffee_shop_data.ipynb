```python
# check out the Normalization of Numerical Data through box plot only unit price variale

sns.boxplot([df["unit_price"], df["Total_Bill"]])
plt.show()

```

```python

# Calculate the IQR

Q1 = df["unit_price"].quantile(0.25)
Q3 = df["unit_price"].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# define the lower bound and upper bownd 


df = df[~((df["unit_price"] < (Q1 - 1.5 * IQR)) | (df["unit_price"] > (Q3 + 1.5 * IQR)))]
```

```python
df.shape
```

```python

sns.boxplot(df["unit_price"])
```

```python
Q1 = df["Total_Bill"].quantile(0.25)
Q3 = df["Total_Bill"].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# Remove the outlier

df = df[~((df["Total_Bill"] < (Q1 - 1.5 * IQR)) | (df["Total_Bill"] > (Q3 + 1.5 * IQR)))]
```

```python
df.shape
```

```python
sns.boxplot(df["Total_Bill"])
```

```python

sns.histplot(data=df, x="unit_price", kde=True)

plt.show()
```
